```python
class KernelLogisticBinaryClassifierGD:
    def __init__(self, degree=2, gamma=1.0, coef0=1.0, lambda_=0.1, lr=0.01, max_iter=10000):
        self.degree = degree
        self.gamma = gamma
        self.coef0 = coef0
        self.lambda_ = lambda_
        self.lr = lr
        self.max_iter = max_iter

    def kernel(self, X1, X2):
        return (self.gamma * X1 @ X2.T + self.coef0) ** self.degree

    def fit(self, X, y):
        y = np.where(y == 1, 1, -1)
        n = X.shape[0]
        self.X_train = X
        K = self.kernel(X, X)
        self.alpha = np.zeros(n)

        for _ in range(self.max_iter):
            f = K @ self.alpha
            p = 1 / (1 + np.exp(y * f))  # logistic probabilities
            grad = -K.T @ (y * p) + self.lambda_ * self.alpha
            self.alpha -= self.lr * grad

    def predict_proba(self, X):
        K_test = self.kernel(X, self.X_train)
        scores = K_test @ self.alpha
        return 1 / (1 + np.exp(-scores))

    def predict(self, X):
        return (self.predict_proba(X) > 0.5).astype(int)
```

ECOC:

```python
class ECOCClassifier:
    def __init__(self, code_matrix, **kernel_params):
        self.code_matrix = code_matrix
        self.kernel_params = kernel_params
        self.classifiers = []

    def fit(self, X, y):
        n_classes, n_classifiers = self.code_matrix.shape
        self.classifiers = []

        for j in range(n_classifiers):
            binary_y = np.array([self.code_matrix[label, j] for label in y])
            clf = KernelLogisticBinaryClassifierGD(**self.kernel_params)
            clf.fit(X, binary_y)
            self.classifiers.append(clf)

    def predict(self, X):
        predictions = np.array([clf.predict(X) for clf in self.classifiers]).T
        pred_classes = []
        for row in predictions:
            distances = np.sum(row != self.code_matrix, axis=1)
            pred_classes.append(np.argmin(distances))
        return np.array(pred_classes)
```

ONE VS ALL:

```python
class OneVsAllKernelLogistic:
    def __init__(self, **kwargs):
        self.base_kwargs = kwargs
        self.models = []
        self.classes_ = None

    def fit(self, X, y):
        self.classes_ = np.unique(y)
        self.models = []
        for c in self.classes_:
            y_bin = (y == c).astype(int)
            model = KernelLogisticBinaryClassifierGD(**self.base_kwargs)
            model.fit(X, y_bin)
            self.models.append(model)

    def predict(self, X):
        probas = np.array([model.predict_proba(X) for model in self.models])
        preds = np.argmax(probas, axis=0)
        return self.classes_[preds]
```

MINI-BATCH:

```python
import numpy as np

class KernelLBClassifierMB:
    def __init__(self, degree=2, gamma=1.0, coef0=1.0,lambda_=0.1, lr=0.01, max_iter=10000, batch_size=None):
        self.degree = degree
        self.gamma = gamma
        self.coef0 = coef0
        self.lambda_ = lambda_
        self.lr = lr
        self.max_iter = max_iter
        self.batch_size = batch_size  # None means full-batch

    def kernel(self, X1, X2):
        return (self.gamma * X1 @ X2.T + self.coef0) ** self.degree

    def fit(self, X, y):
        y = np.where(y == 1, 1, -1)
        self.X_train = X
        self.y_train = y
        n = len(y)
        K = self.kernel(X, X)
        self.alpha = np.zeros(n)

        for _ in range(self.max_iter):
            # Mini-batch indices
            if self.batch_size is None:
                idx = np.arange(n)
            else:
                idx = np.random.choice(n, self.batch_size, replace=False)

            K_batch = K[idx][:, idx]        # K subset for the batch
            alpha_batch = self.alpha[idx]
            y_batch = y[idx]

            f = K_batch @ alpha_batch
            p = 1 / (1 + np.exp(y_batch * f))  # sigmoid(-y*f)
            grad = -K[idx][:, idx].T @ (y_batch * p) + self.lambda_ * alpha_batch
            self.alpha[idx] -= self.lr * grad

    def predict_proba(self, X):
        K_test = self.kernel(X, self.X_train)
        scores = K_test @ self.alpha
        return 1 / (1 + np.exp(-scores))

    def predict(self, X):
        return (self.predict_proba(X) > 0.5).astype(int)
```

```python
class ECOCClassifierMB:
    def __init__(self, code_matrix, **kernel_params):
        self.code_matrix = code_matrix
        self.kernel_params = kernel_params
        self.classifiers = []

    def fit(self, X, y):
        n_classes, n_classifiers = self.code_matrix.shape
        self.classifiers = []

        for j in range(n_classifiers):
            binary_y = np.array([self.code_matrix[label, j] for label in y])
            clf = KernelLBClassifierMB(**self.kernel_params)
            clf.fit(X, binary_y)
            self.classifiers.append(clf)

    def predict(self, X):
        predictions = np.array([clf.predict(X) for clf in self.classifiers]).T
        pred_classes = []
        for row in predictions:
            distances = np.sum(row != self.code_matrix, axis=1)
            pred_classes.append(np.argmin(distances))
        return np.array(pred_classes)
```
